```python
from textblob import TextBlob
import re

def senti_analyze(word):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(word)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
import numpy as np
```

```python
amazon_reviews['Sentiment_Polarity'] = np.array([ senti_analyze(word) for word in amazon_reviews['Reviews'] ])
```

```python
amazon_reviews.head(20)
```

```python
# Constructing lists with classified reviews:

positive_revs = [ review for index, review in enumerate(amazon_reviews['Reviews']) if amazon_reviews['Sentiment_Polarity'][index] > 0]
neutral_revs = [ review for index, review in enumerate(amazon_reviews['Reviews']) if amazon_reviews['Sentiment_Polarity'][index] == 0]
negative_revs = [ review for index, review in enumerate(amazon_reviews['Reviews']) if amazon_reviews['Sentiment_Polarity'][index] < 0]
```

```python
# Printing percentages:

print("Percentage of positive reviews: {}%".format(len(positive_revs)*100/len(amazon_reviews['Reviews'])))
print("Percentage of neutral reviews: {}%".format(len(neutral_revs)*100/len(amazon_reviews['Reviews'])))
print("Percentage of negative reviews: {}%".format(len(negative_revs)*100/len(amazon_reviews['Reviews'])))
```
